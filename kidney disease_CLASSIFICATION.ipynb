```python
# CHANGE THE DATA TYPE OBJECT TO NUMERIC


df.packed_cel_volume=pd.to_numeric(df.packed_cel_volume,errors="coerce")


df.wbc_count=pd.to_numeric(df.wbc_count,errors="coerce")

df.rbc_count=pd.to_numeric(df.rbc_count,errors="coerce")
```

```python
df.info()
```

```python
def mean_value(df,column):
    mean=df[column].mean()
    df[column]=df[column].fillna(value=mean)



def mode_value(df,column):
    mode=df[column].mode()[0]
    df[column]=df[column].fillna(mode)
```

```python
#seperating numerical columns

num_cols = [col for col in df.columns if df[col].dtype!="object"]
for col_name in num_cols:
    mean_value(df,col_name)


#seperating catogorical column

cat_cols=[col for col in df.columns if df[col].dtype=="object"]
for col_name in cat_cols:
    mode_value(df,col_name)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
import numpy as np

param_dist = {
    'n_estimators': [200, 300,400,500],
    'max_depth': [35,40,45,50],
    'min_samples_split': [10,15,20,25],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt'],
    'bootstrap': [True, False]
}

model = RandomForestClassifier(random_state=42)

random_search = RandomizedSearchCV(
    model,
    param_distributions=param_dist,
    n_iter=30,
    scoring='accuracy',
    cv=5,
    verbose=1,
    n_jobs=-1,
    random_state=42
)

random_search.fit(x_train, y_train)
```
